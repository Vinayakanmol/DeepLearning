```python
import spacy 
from spacy import displacy
import pandas as pd
```

```python
df=pd.read_csv('/content/apple.txt',error_bad_lines=False)
```

```python
df
```

```python
df['x']
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
doc=nlp('Mr.Elon is going to acquire Twitter Inc of $45 billion in 2023')
```

```python
doc.ents
```

```python
for i in doc.ents:
  print(i,'|',i.label_)
```

```python
doc1=nlp('Elon is going to acquire Twitter Inc of $45 billion in 2023')
```

```python
for i in doc1.ents:
  print(i,'|',i.label_)
```

##Spacy library is not 100% precise

```python
displacy.render(doc,style='ent',jupyter=True)
```

```python
txt=' '.join(df['x'])
txt
```

```python
x=nlp(txt)
```

```python
displacy.render(x,style='ent',jupyter=True)
```

```python
for i in x:
  print(i,i.pos_)
```
